```python
from scapy.all import *
```

```python
def process_packet(p):
    ##print("******* Packet Information ********")
    ##print( p.summary())
    if p.haslayer(TCP) and p.haslayer(Raw):
        print('\n\nCame From: ', p[IP].src ) ##p.summary())
        print('Going to : ', p[IP].dst )
        print(' \nPAYLOAD : ', p[Raw].load.decode('UTF-8') )
```

```python
packets = sniff( prn =process_packet, filter = 'port 80')
```
